LinkedList are non-contiguous data structure unlike array
They store the values in random amnner and the values are connected via pointers
Since arrays are stored in a contigous manner insertion and deletion operations in array are O(n) oeprations

But in a linkedlist if the insertion and deletion are at the front or the tail we can use the pointers to make those operation in O(1)

LinkedList are used to implement stacks and queues

In a queue a pointer can be at the start and end and FIFO operations can be performed at O(1) time

In a Stack if we have the pointed at the head insertion and deetion can be done in O(1) time

Array are faster when we need to look up a value at a specific given index, in that case values can be found in O(1) space but lookup in a linkedlist is O(n)


# There are two types of linkedlist

Singly linked List 

Doubly Linked List


## Singly Linked List

Singly linked list traverse the list in one direction only
one node is connected to the next node via poinyers

## Doubly Linked List
Doubly linkedlist can traverse ibn both directions
One node is connected to the previous node and next node

one benefit of doubly linkedlist is if you want to look at the previous element you can do that in O(1) but in a singly linked list you have to start from head and that will be O(1)
