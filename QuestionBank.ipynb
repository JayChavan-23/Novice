## What is canonicalisation ? What is an example of how it can be applied

In simple words canonicalisation is the process of data that can have many different form and forcing it into a single standard form

FOr example
Common Example: Web Addresses (URLs)
Canonicalization is essential on the internet:

Problem: The following two web addresses lead to the exact same home page:

http://example.com

http://www.example.com

Canonical Solution: The web server chooses one as the canonical version (e.g., the non-www version). When a browser tries to access the www version, the server automatically redirects it to example.com.

Benefit: This prevents search engines from seeing two different pages with the same content, which avoids splitting the page's ranking power (called "link authority") and ensures users always land on the preferred address.
