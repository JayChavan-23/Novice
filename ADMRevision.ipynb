Explain the difference between algorithms and heuristic

An Algorithm is a set of instruction/procedure we follow to achieve a particular task and there’s a guarantee that algorithm is correct and it could be optimal depends on the way you implement it 
While correct algorithm for certain complex problems can take a impractical amount of time

Heuristic is like a general rule of thumb approach, is is like an educated guess, it might not necessarily be the optimal or correct solution but it saves time and computational efforts. 
For example Amount of baking time for a cake with no recipe based on previous experience it is a heuristic approach it’s fast but the result might not be perfect

Why always Big Oh?
Big O works on the upper bound of an. Algorithm growth rate this means it tells you the maximum time a algorithm might take for a particular input size which is crucial when calculating worst case time complexities for algorithms in computer science

Big Omega works on the lower bound which always gives us the best case time of an algorithm.

Big theta gives the tight bound that means the best and worst case scenario are the same

## Dict and Hash Tables

Dictionaries are generic way of mapping values in key value pairs

Hashtable is a specific way of representing a dict using a hash function

What is a hash Function?
It is a function that takes in a key and then converts it into a int which is a key where the value is stored

Now it is possible that hasing two different values can result in the same index which is called as collison 

To overcome collision we introduce chaining

We use linkedlist to store values at that index

Key properties a hash function should have 
- Produce least collisions

Popular ways to design a hash function

- Division - k mod m where k is the key and m is the table size
- Multiplacation
- Mid swaure method - square the key and then take the middle digits =(or bits) as the hash value

Universal Hashing
-Pick a hash function at random from a family of hash functions to reduce the chance of worst-case collisions.
