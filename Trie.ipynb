## Trie -> special type of tree but it is mainly used to store characters

- Each node represnerts a character
- Each Path from root to a node forms a prefix
- Each complete word ends at a node marked as “end of word”

Trie has fast prefix search
- prefix is a starting part of words
- The time to find a prefix depends on the length of the word not the number of words present in the trie

TIme complexity

L - > Length of Word

Insert , Search and Prefix Search -> O(L)



Application of trie

- Autocomplete 
- spellcheck
- Word suggestions
