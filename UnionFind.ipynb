## Union Find also called as Disjoint Set Union

It keeps track of multiple disjoint sets and suppports 2 operations
- makeset (Makes a new set)
- find (x) - Finds the parent of the set contaning x
- Union (x,y) - > Merges the sets containign x and y

Application
- It helps to detect uf a graph has cycle in undirected graphs
- which nodes re connected in a network (Network connectivity)
- Minimum sopanning tree (Kruskal's Algo)
