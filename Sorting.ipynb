Naive sorting algorithm were O(n^2) but now new algorithms like Quick sort and Merge sort Run O(n log n) time

## Merge Sort
- Follows Divide and conquer strategy (Divides the array from the center or nearly center)
- It recursively divides the array into two halves
- and then merginig it back together to obtain the sorted order

Time Complexcity
- Best Case: O(n logn): When the array is sorted or nearly sorted
- Average Case: O(n logn): When the array is randomly ordered
- Worst Case: O(n logn): When the array is sorted in reverse order

Pro's 
- Merge Sort is a Stable sorting algorithm [It always take the element from the left side first hence the order is preserved]


## Quick Sort
- Follows Divide and conquer (Divides the array from a pivot element) Pivot element can be first element, last element, random element, median of three
- partition the array such that elements smaller than the pivot go to the left, greater than the pivot go to the right

Time Complexity
- Best Case: O(n logn)
- Average Case: O(n logn)
- Worst Case: O(n^2)

Cons
- Quick Sort is not a stable algorithm because swaps during partitioning may chang the order of the equal elements


## Why is Quick sort still preferred over merge sort?

Quick Sort is fasters in real life
Merge sort requires O(n) space but quicksort can be done in-place

Even tho worst case is O(n^2) pivot selection techniques like random pivot or median of three pivot selection the worst case is avoided

Greater Cache Performance

## What is a Stable Sorting Algorithm?

Stable sorting algorithm is that which preserves the order if the equal value elements

Stable Sorting Algorithms
- Merge Sort
- Insertion Sort 
- Bubble Sort
- Counting Sort
- Radix Sort
- Tim Sort

Unstable Sorting Algorithms
- Quick Sort
- Heap Sort
- Selection Sort

## Heap Sort
Heap is a ordered binary tree
Max heap is where the value of the parent node is greater than the value of child node (parent > child)
The element at the top of the max heap (root element) is the biggest element

build-max-heap: Creates a max heap from an unsorted array
Heapify: Similar to build max heap but assumes part of array is already sorted

Steps to acheive a heap sort
- create a max heap
- remove the largest item
- place then in the sorted array

Time Complexity O(n logn) all cases
- Build max heap takes O(n)
- Heapify: O(logn)

## Selection Sort
Looks through the array over and over and moving the next lowest value to the sorted part O(n^2)

## Insertion sort
Takes each element compares it to the left side(sorted) side of the array and puts it in the right position

Best case O(n)  Given that array is already sorted
Avg and worst case O(n^2)

## Counting Sort
Counts how many time the value has appeared and then uses these count to place item in the correct position
Time Complxity in all cases is going to be O(n+k) given that n is the number of eleement and k is the range of values (0 - k-1)

## Radix Sort
At each level we pick a base like log 10 and we perform counting sort d amount of time where d is the length of the digit

Hence the time complexty is 

O(d . (n+k))

d is the digits
= is the base and n is the number of elements

## Bubble sort
Repeatedly compares adjacent elements and swaps them it they are in th wrong order

Large elements bubble up to the end of the array

Time Complexity
Best O(n) When array is sorted
Avg and Best case O(n^2)

## Bucket Sort
Divides the range into equal sized buckets

Distributes elements into these buckets based on the value

Sort each bucket (ususally via insertion sort)

Concatenate all buckets

Time Complexity
Best O(n)

Average Case O(n+k) n is the number of element and k is the range of elements

Worst Case  n^2 when all the elements fall in the same bucket
