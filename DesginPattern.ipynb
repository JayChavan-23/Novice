An object is something that stores data and can perform actions.

## Singleton Design Patter

Only One object of a class exist and that object can be accessed anyhwhere in the code

Real life example;- A country can have only one president
Everyone referes to the preseident to a single person



## Factory Design Pattern

Multiple Objects are created to not expose the creation logic

Real-World Analogy

Imagine a pizza store.

You don’t go to the kitchen and make your own pizza.
You tell the Pizza Factory:

"I want a pepperoni pizza."


The factory returns the correct object:

PepperoniPizza

VegPizza

CheesePizza

You don’t know the recipe → creation logic is hidden.

| Feature                | **Singleton Pattern**                      | **Factory Pattern**                                               |
| ---------------------- | ------------------------------------------ | ----------------------------------------------------------------- |
| Purpose                | Ensure **only one instance**               | **Create many different objects** without exposing creation logic |
| Objects created        | Only **one**                               | **Many**                                                          |
| Who controls creation? | The class itself                           | The factory class                                                 |
| When used?             | When you need a **single shared resource** | When object creation is complex or dynamic                        |
| Example                | Logger, Configuration, DB connection       | ShapeFactory, PizzaFactory, PaymentFactory                        |
| Constructor            | **Private**                                | Usually **public**                                                |
| Global access?         | Yes, via `getInstance()`                   | No, factory simply returns objects                                |
| Problem solved         | Avoids duplication                         | Avoids complex `new` logic everywhere                             |
| Pattern type           | Creational                                 | Creational                                                        |
