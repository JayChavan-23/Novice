## Depth First search

Time complexity: O(V+E)

v = Visiting every vertex | E - For every vertex we we explore atmost 3 edges

Space Complexity

O(v) the graph could be like a skewed linkedlist in that case 

## Breadth Fiest Search

Time Complexity: O(v + E) when using Adjacency list

O(v^2) when using adjacency matrix because for every V we are scanning the entire row

Space is O(V) we need queue
