## Minimum spanning trees

MST are subset of edges of a connected wieighted, undirected graph which conects all the edges with minumum weight

Connected - so that we can travel from one node to another and every vertex should be reachable

Weighted so that we can calculate the minimum weight of the edges

undirected - so that we can travel both ways

## How to make a Minimum Spanning tree


## Kruskal's Algorithms

Via Kurskal's algorithm we can find the minimum spanning tree of a connected weighted undirected graph

Step 1: First we sort the edges to find the lowest weight of the edges with usually a sorting algorithm like quick sort or merge sort

then we start with the lowest/cheapest edge and travel to another edge

we make sure that the new edge does not form a cycle with the help of DSU (if they are in the same set it forms a cycle we have to make sure the new edge does not belong in the same set)

if it forms a cycle then we skip it and move on to next edge

once the spanning tree is complete we notice that the number of edges is V-1

Time complexity is E log E

It's a mathematical term that a conencted graph that has vertex V andno cycles must have v-1 edges

Fewer edges then it wont be connected | more edges then it will have a cycle

## Prim's Algorithm

Prims is another way of finding the MST of a connected weighted undirected graph, 

Start from any vertex (Unlike Kruskals)

Now at each step pick the cheapest  (We cna use a priority queue[min heap] to get the cheapest edge) O(1)

Then we continue this process until we have v-1 edges

We maintain 2 sets
- visited (MST set): to track which vertex has been included in the MST
- Unvisited (Usually a Priority queue) - to effectively retrieve the cheapest edge that connects a vertex
