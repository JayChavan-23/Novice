## Binary Search Tree

Binary search tree is basically a complete or a perfect binary tree which has a property that all the nodes on the left side of a parent node whill be smaller than the parent node and ll the ndeps on the right of the bianry tree will be greater than the parent 


Lookup -> log n

It's good to have it balanced or else it will look like a linkedlist and it will take worst case lookup O(n)

If the Binary search tree is Height balanced then the lookup would be O(logn) but if it is not then worst case would be O(n)

A cool property about binary search tree, if you traverse the elements via in a in-order traversal method they will be in a sorted order.
