## Quick Select algorithm

It is usually used to find the kth smallest or the kth largest element in an unsorted array

it is based on same idea as quick sort but instead of sorting the whole array it recursively explores the side where the kth element is


Pick a pivot (usually random)

Partition the array into:

Left side → elements smaller than pivot

Right side → elements greater than pivot

Compare the pivot's final index with k

Only recurse on the side that contains the k-th smallest element

This is like QuickSort, but only goes down one branch → that's why it’s O(n).

Best / Average Case: O(n)

Worst Case: O(n²) (rare with random pivots)


Why QuickSelect Works Fast

QuickSelect eliminates half the array each step (on average).

Unlike sorting the whole array:

Sorting → O(n log n)

QuickSelect → O(n)

Because QuickSelect doesn’t sort both sides of the pivot —
only the side containing the k-th element.
