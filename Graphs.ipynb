## Graphs
Vertices [v] -> to store data

Edges [e] -> to link data





Directed Graph: It has a direction that it follows

Undirected graph: It can ve bidirectional

Cyclic Graph: There is a infinite loop in the graph

Adjacency Matrix: 2-d list of nxn rows and colums 

Rows is from -> Column is to

Adjacency List: [It's a hashmap]

Key as the node and it's value is all the connection it leads to like {0:[1,3], 1:[2], 2:[], 3:[4,6,7]}

```python
#Array of edges (Directed) [Start,End]

n = 8
A = [[0, 1], [1, 2], [0, 3], [3, 4], [3, 6], [3, 7], [4, 2], [4, 5], [5, 2]]

A
```

```python
# COnverting an Array of edges to Adjacency Matrix

M = []
for i in range(n):
  M.append([0]*n)

for u,v in A:
  M[u][v] =1
M
```

```python
# Converting array of edges to a adjacency list
from collections import defaultdict

D = defaultdict(list)

for u,v in A:
  D[u].append(v)

D
```

## Depth First Search [DFS] -> Stack [Vertically] O(v+e)

Picks one side and explores the same path until hit a dead end and then traverses back with backtracking

We maintain a seen set (hashset) to make sure we are not going to the same node again, as we move along we keep adding elements to the seen set

because we have a seen set this avoids us looping over and over in cycles







```python
# DFS with Recursion

def dfs_recursive(node):
  print(node)
  for nei_node in D[node]:
    if nei_node not in seen:
      seen.add(nei_node)
      dfs_recursive(nei_node)


source = 0
seen = set()
seen.add(source)
dfs_recursive(source)
```

## Breadth First Search [BFS] -> Queue [Horizontally] O(v+e)





# BFS with Queue While loop

source = 0
from collections import deque

seen = set()
seen.add(source)

q.append(source)
